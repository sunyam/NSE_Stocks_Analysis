```python
# Analysing Risk
# Examining the Expected Daily Return (mean) vs Risk involved (deviation in the daily returns)

DailyReturns = DailyReturns.dropna()

plt.scatter(DailyReturns.mean(), DailyReturns.std(), s=10)
plt.xlabel('Expected Return (mean)')
plt.ylabel('Risk Involved (std.)')

for columnName, x, y in zip(DailyReturns.columns, DailyReturns.mean(), DailyReturns.std()):
    plt.annotate(columnName[0], xy=(x, y))    # Representing a stock with the intitial letter

# Ideally, the stock should be towards the lower right- least risk with maximum returns.
# If the returns and risk are both high, then you are really 'playing' the stock market.
# All the stocks except Kaveri Seeds have negative expected returns. Not good.
# Wockhardt is the worst one to invest in (high risk and least expected returns)
```

Monte Carlo Simulation:

- Performing Monte Carlo simulation for the next 1 year to get a sense of the stock's behaviour in the future.

- The concept and equations for the Monte Carlo method have been taken from this great investopedia article:
http://www.investopedia.com/articles/07/montecarlo.asp

I highly recommend you read this article if you are not familiar with terms like "drift", "shock".

```python
numberOfDays = 365
dt = 1/numberOfDays

# Drift is given by 'dt' times 'mu' (expected return)
mu = DailyReturns['K_DR'].mean()

# Shock depends on the standard deviation of daily returns (sigma)
sigma = DailyReturns['K_DR'].std()

from __future__ import division

# Main function: It will return an array of simulated prices for the next one year (sim_price)
def monte_carlo(initialPrice):
    
    # Initialise arrays
    sim_price = np.zeros(numberOfDays)
    drift = np.zeros(numberOfDays)
    shock = np.zeros(numberOfDays)
    
    # Give the simulation a starting point
    sim_price[0] = initialPrice

    # Need to find prices for each day
    for i in xrange(1, numberOfDays):
        
        # Calculating Drift:
        drift[i] = mu * dt
        
        # Calculating Shock:
        root_dt = np.sqrt(dt)
        scale = sigma * root_dt
        shock[i] = np.random.normal(loc=mu*dt, scale=scale)
        
        # Calculating simulated price:
        # Referring to the change in price as 'delta_price'
        delta_price = sim_price[i-1] * (drift[i] + shock[i])
        sim_price[i] = sim_price[i-1] + delta_price

    return sim_price
```

```python
# Getting the initial-price
KaveriSeed.tail()
```

```python
initialPrice = 455.0

plt.figure(figsize=(16,6))
for i in xrange(100):
    plt.plot(monte_carlo(initialPrice))

plt.xlim(0, 370)
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Monte Carlo Simulation for Kaveri Seed Company Ltd.')
```

```python
# Plotting a Histogram for a simulation of a larger number of runs
# Only concerned with the price at the 365th day now

runs = 10000

simulations = np.zeros(runs)

for run in xrange(runs):
    simulations[run] = monte_carlo(initialPrice)[numberOfDays-1]
```

```python
plt.figure(figsize=(16,6))
plt.hist(simulations, bins=200)
plt.title("Price Distribution for Kaveri Seed Company Ltd. after " + str(numberOfDays) + " days")
```
